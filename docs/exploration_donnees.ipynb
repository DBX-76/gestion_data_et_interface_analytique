```python
import pandas as pd

# --- 1. Force les dates critiques ---
date_cols_ops = ['date_heure_reception_alerte', 'date_heure_fin_operation']
for col in date_cols_ops:
    if col in operations_df.columns:
        operations_df[col] = pd.to_datetime(operations_df[col], errors='coerce')

if 'date' in operations_stats_df.columns:
    operations_stats_df['date'] = pd.to_datetime(operations_stats_df['date'], errors='coerce')

# --- 2. Vérifications par table ---

print("🔍 1. OPERATIONS – Catégories suspectes")
cat_cols_ops = ['type_operation', 'pourquoi_alerte', 'moyen_alerte', 'qui_alerte', 'categorie_qui_alerte', 'cross', 'evenement', 'vent_direction_categorie']
for col in cat_cols_ops:
    if col in operations_df.columns:
        print(f"\n{col}:")
        print(operations_df[col].value_counts(dropna=False).head(10))

print("\n⚠️ 2. OPERATIONS – Vérifications numériques & dates")
# Dates futures ?
today = pd.Timestamp.today(tz='UTC')
future_start = operations_df[operations_df['date_heure_reception_alerte'] > today]
future_end = operations_df[operations_df['date_heure_fin_operation'] > today]
if len(future_start) > 0:
    print(f"• Date réception dans le futur : {len(future_start)}")
if len(future_end) > 0:
    print(f"• Date fin dans le futur : {len(future_end)}")

# Durée négative ?
operations_df['duree_sec'] = (operations_df['date_heure_fin_operation'] - operations_df['date_heure_reception_alerte']).dt.total_seconds()
neg_duration = operations_df[operations_df['duree_sec'] < 0]
if len(neg_duration) > 0:
    print(f"• Opérations avec durée négative : {len(neg_duration)}")

print("\n🔍 3. FLOTTEURS – Valeurs catégorielles")
for col in ['pavillon', 'resultat_flotteur', 'type_flotteur', 'categorie_flotteur']:
    if col in flotteurs_df.columns:
        print(f"\n{col}:")
        print(flotteurs_df[col].value_counts(dropna=False).head(10))

print("\n🔍 4. RÉSULTATS HUMAINS – Valeurs catégorielles")
for col in ['categorie_personne', 'resultat_humain']:
    if col in resultats_humain_df.columns:
        print(f"\n{col}:")
        print(resultats_humain_df[col].value_counts(dropna=False).head(10))

print("\n🚨 5. INCOHÉRENCES MÉTIER")

# Règle 1 : opération terminée → doit avoir date_fin
term_sans_fin = operations_df[
    operations_df['date_heure_reception_alerte'].notna() &
    operations_df['date_heure_fin_operation'].isna()
]
if len(term_sans_fin) > 0:
    print(f"• Opérations avec alerte mais pas de fin : {len(term_sans_fin)}")

# Règle 2 : si "sans_flotteur_implique" = True, alors flotteurs_df ne devrait pas avoir de ligne
if 'sans_flotteur_implique' in operations_stats_df.columns:
    ops_sans_flotteur = operations_stats_df[operations_stats_df['sans_flotteur_implique'] == True]['operation_id']
    conflit = flotteurs_df[flotteurs_df['operation_id'].isin(ops_sans_flotteur)]
    if len(conflit) > 0:
        print(f"• Conflit : opérations marquées 'sans_flotteur' mais présentes dans flotteurs.csv : {len(conflit)}")

# Règle 3 : résultat humain "tiré d'affaire" mais nombre = 0 ?
humain_zero = resultats_humain_df[resultats_humain_df['nombre'] == 0]
if len(humain_zero) > 0:
    print(f"• Lignes résultats humains avec nombre = 0 : {len(humain_zero)}")
```
